### Function Graph Visualization

```python
import numpy as np
import matplotlib.pyplot as plt


# generate an array of values
# of the independent variable
x = np.linspace(0, 1, 5000)

ax = plt.gca()

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')  

# set range of the horizontal axis
plt.xlim([0, 1])    
# set range of the vertical axis
plt.ylim([-0.25, 0.25]) 

# plot the right side
plt.plot(x, (3 * x - 1) / (4 * x * x - 4 * x + 17))

# show the plot
plt.show()
```

### Numerical Integation with Scipy

#### quad()

```python
import scipy
from scipy.integrate import quad

def f(x):
    return (3 * x - 1) / (4 * x * x - 4 * x + 17)

# quad() is a function that calculate definite integrals.
# It returns an evaluated value of the integral and estimated absolute error.
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.quad.html
I = quad(f, 0, 1)

# The output contains the value of the definite integral and the estimated precision
print(I)
```

#### trapz()

```python
from numpy import *
from scipy.integrate import trapz

def f(x):
    return (3 * x - 1) / (4 * x * x - 4 * x + 17)

a = 0
b = 1

n = 300
h = (b - a) / n

x = arange(a, b + h, h)
#print(x)

y = []
for xi in x:
    y.append(f(xi))

#print(y)

I = trapz(y, x = x)
print(I)
```
